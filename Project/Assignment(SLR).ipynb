Canada Per Capita Income

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
df = pd.read_csv(r"C:\Users\Lenovo\Downloads\canada_per_capita_income.csv")
df.head(5)
```

```python
plt.xlabel('Year')
plt.ylabel('Income')
plt.scatter(df['year'],df['per capita income (US$)'])
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
#reg.predict(np.array([[2020]], dtype=np.float64))

X_new = pd.DataFrame([[2020]], columns=['year'])
reg.predict(X_new)
```

```python
plt.xlabel('Year')
plt.ylabel('Income')
plt.scatter(df['year'],df['per capita income (US$)'])
plt.plot(df.year,reg.predict(df[['year']]),color = 'green')
plt.show()
```
